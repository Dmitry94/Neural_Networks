```python
import time
import numpy as np

import utils
import LinearSVM as lSVM

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Generate data

X, Y = utils.generate_spiral_data(100, 3)
X_test, Y_test = utils.generate_spiral_data(300, 3)
```

```python
# Visualize spirals

plt.scatter(X[:, 0], X[:, 1], c=Y, s=40, cmap=plt.cm.Spectral)
plt.figure(figsize=(300, 200))
plt.show()
```

```python
# Init classifier

BATCH_SIZE = 512
LEARNING_RATE = 0.01
REG_LAMBDA = 0.01

classifier = lSVM.SVMLinearClassifier(learning_rate=LEARNING_RATE, batch_size=BATCH_SIZE, reg_lambda=REG_LAMBDA)
```

```python
# Training

start = time.clock()
classifier.train(X, Y)
end = time.clock()
print 'SVM Training time = ', end - start
```

```python
# Predicting

start = time.clock()
predictions = classifier.predict(X_test)
end = time.clock()
print 'SVM Predicting time = ', end - start

accuracy = np.mean(predictions == Y_test)
print 'SVM Accuracy = %f' % accuracy
```

```python
# Visualize spirals

plt.scatter(X_test[:, 0], X_test[:, 1], c=predictions, s=40, cmap=plt.cm.Spectral)
plt.figure(figsize=(300, 200))
plt.show()
```
