```python
# Visualize spirals

plt.scatter(X[:, 0], X[:, 1], c=Y, s=40, cmap=plt.cm.Spectral)
plt.figure(figsize=(300, 200))
plt.show()
```

```python
# Init classifier

LEARNING_RATE = 1e-2
REG_LAMBDA = 1e-3

svm = lSVM.SVMLinearClassifier(learning_rate=LEARNING_RATE, reg_lambda=REG_LAMBDA)
softmax = sfmx.SoftmaxClassifier(learning_rate=LEARNING_RATE, reg_lambda=REG_LAMBDA)

HIDDEN_LAYERS_SIZES = [50, 75]
net = BNN.BNN(HIDDEN_LAYERS_SIZES)
```

```python
# Training

start = time.clock()
svm.train(X, Y, max_iters=1200)
end = time.clock()
print 'SVM Training time = ', end - start

start = time.clock()
softmax.train(X, Y, max_iters=1200)
end = time.clock()
print 'Softmax Training time = ', end - start

start = time.clock()
net.train(X, Y, print_loss=True)
end = time.clock()
print 'Net Training time = ', end - start
```

```python
# Predicting

start = time.clock()
svm_predictions = svm.predict(X_test)
end = time.clock()
print 'SVM Predicting time = ', end - start

accuracy = np.mean(svm_predictions == Y_test)
print 'SVM Accuracy = %f' % accuracy


start = time.clock()
softmax_predictions = softmax.predict(X_test)
end = time.clock()
print 'Softmax Predicting time = ', end - start

accuracy = np.mean(softmax_predictions == Y_test)
print 'Softmax Accuracy = %f' % accuracy


start = time.clock()
net_predictions = net.predict(X_test)
end = time.clock()
print 'Net Predicting time = ', end - start

accuracy = np.mean(net_predictions == Y_test)
print 'Net Accuracy = %f' % accuracy
```

```python
# Visualize spirals

plt.scatter(X_test[:, 0], X_test[:, 1], c=softmax_predictions, s=40, cmap=plt.cm.Spectral)
plt.figure(figsize=(300, 200))
plt.show()

plt.scatter(X_test[:, 0], X_test[:, 1], c=svm_predictions, s=40, cmap=plt.cm.Spectral)
plt.figure(figsize=(300, 200))
plt.show()

plt.scatter(X_test[:, 0], X_test[:, 1], c=net_predictions, s=40, cmap=plt.cm.Spectral)
plt.figure(figsize=(300, 200))
plt.show()
```
